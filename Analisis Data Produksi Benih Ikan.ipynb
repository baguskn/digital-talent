```python
import pandas as pd
data_benih = pd.read_csv('jumlahproduksibenih.csv', delimiter = ',', encoding = "ISO-8859-1")
print("Data Jumlah Produksi Benih Ikan Nasional Tahun 2001-2012")
data_benih
```

# 1. Deskripsi Data 
Data Jumlah Produksi Benih Ikan Nasional Tahun 2001-2012 adalah data yang berisi tentang jumlah produksi benih ikan yang diproduksi di Republik Indonesia dalam rentang waktu tahun 2001-2012 berdasarkan jenis benih dan cara budidaya.

```python
# Fungsi untuk menampilkan object kolom
def objek_kolom(list,n):
    z=''
    for i in range(0,n):
        if i!=n-1:
            z=z+list[i]+', '
        else: 
            z=z+'dan '+list[i]+'.'
    return z

#1. Menghitung jumlah kolom dan menampilkan object kolom
list_kolom=data_benih.columns
kolom=len(data_benih.columns)
print('Data Memiliki',kolom,'kolom, yaitu:',objek_kolom(list_kolom,kolom))
data_benih.columns
```

# 2. Statistik Data

### Total Produksi Benih Ikan Nasional Selama 2001-2012

```python
total=data_benih.Jumlah.sum()
print('Total Produksi Benih Ikan Nasional Selama 2001-2012:',total)
```

### Total Produksi Benih Ikan Nasional berdasarkan Tahun

```python
data_benih.groupby(['Tahun']).sum()['Jumlah']
```

# 3. Visualisasi Sederhana

```python
ax = data_benih.groupby('Tahun').sum()['Jumlah']
ax = ax.plot(kind='bar', title ="Jumlah Produksi Benih Ikan per Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

# 4. Analisis Data
Melihat visualisasi data di atas, statistik jumlah produksi benih ikan nasional selama tahun 2001-2012 mengalami kenaikan yang sangat pesat dimulai tahun 2004 sampai tahun 2005. Namun, setelah itu mengalami penurunan hampir setengahnya pada tahun 2006 dan 2007, dan penurunan drastis terjadi pada tahun 2008. Produksi benih ikan nasional paling banyak terjadi pada tahun 2005, dan paling sedikit di tahun 2012.
